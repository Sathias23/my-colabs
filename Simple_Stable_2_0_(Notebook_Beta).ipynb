<h2><b>HOW TO USE</b></h2>

1. Scroll down to "Generate!" and select a model under "Model Choice"
2. Put a text prompt in "Prompt:"
3. Press the play button to the left<br/> ![Image](https://cdn.discordapp.com/attachments/1019436359867453520/1060166210064498758/image.png)
4. Simple Stable will take a few minutes to download and set itself up, and then your image will pop out
5. Press play again to generate more images
6. You can change the model for different styles, but each new model you select will take a few minutes to download and set up

  <summary><big><b>Advanced Settings</b></big></summary>

  **Prompt Weights and Scheduling**

  The default prompt is weighted at 1.

  "Increasing attention" means emphasizing the element, so if you want more red in your output, you can use `(red:2), rest of the prompt` to ask the prompt to try to be more red.

  "Decreasing attention" means decreasing the element, so if you want some Mona Lisa but not a lot, you can use `Rest of the prompt, (Mona Lisa:0.5)` to have some but not too much Mona Lisa.

  Technically words at the beginning of the prompt are the strongest, and words at the end of the prompt are the weakest.

  <br/>

  Example | Description
  --- | --- 
  `(the land of frogs:1.5), (a beautiful landscape:0.5)` | `the land of the frogs` is weighted at `1.5` <br /> `a beautiful landscape` is weighted at half, or `0.5`
  `(red), ((blue)), [yellow], [[green]]` | `red` at 1.1 <br/> `blue` at 1.21 (= 1.1 x 1.1) <br />`yellow` at 0.9 (1 divided by 1.1) <br /> `green` at 0.83 ((1/1.1)*(1/1.1))
  `[cat:dog:0.5] in a pond` | `cat in a pond` for the first 50% of the generation <br /> `dog in a pond` for the second 50% of the generation
  `[cat\|dog] in a tree` | every step, flips between `cat in a tree` and `dog in a tree`

  Complex example
  <br/> <sub>`fantasy landscape with a [mountain:lake:0.25] and [an oak:a christmas tree:0.75][ in foreground::0.6][ in background:0.25] [shoddy:masterful:0.5]`</sub>

  <sub>If 100 steps is used <br>
  at start, `fantasy landscape with a mountain and an oak in foreground shoddy` <br/>
  after step 25, `fantasy landscape with a lake and an oak in foreground in background shoddy` <br/>
  after step 50, `fantasy landscape with a lake and an oak in foreground in background masterful`<br/>
  after step 60, `fantasy landscape with a lake and an oak in background masterful` <br/>
  after step 75, `fantasy landscape with a lake and a christmas tree in background masterful` <br/>
